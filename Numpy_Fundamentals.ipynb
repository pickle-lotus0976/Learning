# Array Operations

```python
a1 = np.arange(12).reshape(3,4)
a2 = np.arange(12, 24).reshape(3,4)
a3 = np.random.random((3,4))

print(a1, "\n")
print(a2, "\n")
print(a3, "\n")
```

```python
# Scalar Operation
# Arithmetic Operation
print(a2 ** 2)
print(a1 * 2)
print(a1 + 2)
print(a2 / 2)
print(a2 % 2)
a3 = np.round(a3 * 100)
a3 = a3.astype(np.int32)
print(a3)
```

```python
# Relational Operation
print(a2 < 35)
```

```python
ar1 = np.random.random((3,4)) * 100
ar2 = np.random.random((3,4)) * 100
print(ar1, "\n")
print(ar2)
```

```python
ar1 = np.ceil(ar1)
ar2 = np.floor(ar2)
print(ar1, "\n")
print(ar2)
```

```python
# Mean
av = np.mean(a1)
# Median
mu = np.median(a2)
# Standard Deviation
sd = np.std(a2)
# Variance
v = np.var(a2)
print(av)
print(mu)
print(sd)
print(v)
```

```python
# Dot product
product = np.dot(a1, a2.transpose())
print(product)
```

```python
# Stacking
arr1 = np.arange(12).reshape(3,4)
arr2 = np.arange(12,24).reshape(3,4)
print(arr1)
print(arr2)

# Horizontal Stacking
print(np.hstack((arr1,arr2)))
# Vertical Stacking
print(np.vstack((arr1, arr2)))
# Horizontal Splitting
print(np.hsplit(arr1,2))
# Vertical Spliting
print(np.vsplit(arr2,3))
```
