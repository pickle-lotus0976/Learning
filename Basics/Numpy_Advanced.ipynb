# Numpy Arrays Vs Python Lists


```python
# Speed of execution
# List
import time
a = [i for i in range(10000000)]
b = [i for i in range(10000000, 20000000)]

c = []
start = time.time()
for i in range(len(a)):
  c.append(a[i] + b[i])
print(time.time() - start)
```

```python
# Numpy Array
import numpy as np
import time
a = np.arange(10000000)
b = np.arange(10000000, 20000000)
start = time.time()
c = a + b
print(time.time() - start)
```

```python
# Memory
# Python List
import sys
a = [i for i in range(10000000)]
sys.getsizeof(a)
```

```python
# Numpy Array
import sys
a = np.arange(10000000, dtype=np.int8)
sys.getsizeof(a)
```

# Broadcasting
The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.  Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. Broadcasting provides a means of vectorizing array operations so that looping occurs in C instead of Python. It does this without making needless copies of data and usually leads to efficient algorithm implementations. NumPy operations are usually done on pairs of arrays on an element-by-element basis.

```python
# Same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)
print(a)
print("\n",b,"\n")
print (a + b)
```

```python
# Different Shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)
print(a)
print("\n",b,"\n")
print (a + b)
```

```python
print(np.arange(3).reshape(3,1) + np.arange(3))
```

```python
# Broadcasting Examples
a = np.arange(12).reshape(4,3)
b = np.arange(3)
print(a)
print(b)

print(a + b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(3)
print(a)
print(b)

print(a + b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a + b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(4).reshape(4,1)

print(a)
print(b)

print(a + b)
```

```python
a = np.array([1])
# shape -> (1,1)
b = np.arange(4).reshape(2,2)
# shape -> (2,2)

print(a)
print(b)

print(a + b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)

print(a)
print(b)

print(a + b)
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)

print(a)
print(b)

print(a + b)
```

```python
a = np.arange(10)
np.sin(a)
```

```python
# Sigmoid Function
def sigmoid(arr):
  return 1/(1 + np.exp(-(arr)))
sigmoid(a)
```

```python
# Mean Squared Error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

def mse(a, b):
  return np.mean(np.square(a - b))

mse(actual, predicted)
```

# Graphing and Plotting

```python
# y = xlog(x)
x = np.linspace(-10,10,100)
y = x * (np.log10(x))
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
```

```python
# Scatter plot
# y = xlog(x)
x = np.linspace(-10,10,100)
y = x * (np.log10(x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```
