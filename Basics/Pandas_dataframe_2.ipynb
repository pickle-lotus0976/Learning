```python
# apply (Series + Dataframe)
points_df = pd.DataFrame(
    {
        '1st_point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd_point':[(-3,4),(-0,0),(2,2),(10,10),(1,1)]
    }
)
points_df
```

```python
def euclidean_distance(row):
  pt_A = row['1st_point']
  pt_B = row['2nd_point']
  dist = np.sqrt((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)
  return dist
```

```python
points_df['euclidean_dist'] = points_df.apply(euclidean_distance, axis=1)
points_df
```
