```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import torch
import torch.nn as nn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

# Scaling

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

# From Numpy to Torch Tensors

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

# Defining Model

```python
# class MySimpleNN(nn.Module):
#   def __init__(self, num_features) -> None:
#     super().__init__()
#     self.linear = nn.Linear(num_features, 1)
#     self.sigmoid = nn.Sigmoid()

#   def forward(self, features) -> torch.Tensor:
#     out = self.linear(features)
#     out = self.sigmoid(out)
#     return out

#   def binary_cross_entropy(self, y_pred, y):
#     epsilon = 1e-8
#     y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
#     loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
#     return loss
```

```python
class MySimpleNN(nn.Module):
  def __init__(self, num_features) -> None:
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features) -> torch.Tensor:
    out = self.linear(features)
    out = self.sigmoid(out)
    return out
```

# Parameters

```python
learning_rate = 0.1
epochs = 1000
```

# Calling the loss function directly from nn module

```python
# loss function
loss_function = nn.BCELoss()
```

# Training Pipeline

```python
# model = MySimpleNN(X_train_tensor.shape[1]) # Initialize with X_train_tensor
# model.double() # Cast model parameters to float64
# for epoch in range(epochs): # Changed variable name from epochs to epoch
#   # forward propagation
#   y_pred = model(X_train_tensor)

#   # Loss
#   loss = loss_function(y_pred, y_train_tensor.view(-1,1).double())
#   if epoch % 100 == 0:
#     print(f"Epoch : {epoch}, Loss : {loss.item()}")
#   # Backpropagation
#   loss.backward()

#   # parameters update
#   with torch.no_grad():
#     model.linear.weight -= learning_rate * model.linear.weight.grad
#     model.linear.bias -= learning_rate * model.linear.bias.grad

#   # zero grads
#   model.linear.weight.grad.zero_()
#   model.linear.bias.grad.zero_()
```

```python
model = MySimpleNN(X_train_tensor.shape[1]) # Initialize with X_train_tensor
model.double() # Casting to float64

# optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(epochs): # Changed variable name from epochs to epoch
  # forward propagation
  y_pred = model(X_train_tensor)

  # Loss
  loss = loss_function(y_pred, y_train_tensor.view(-1,1).double())
  if epoch % 100 == 0:
    print(f"Epoch : {epoch}/{epochs}, Loss : {loss.item()}")

  # clearing the gradients
  optimizer.zero_grad()

  # Backpropagation
  loss.backward()

  # Updating parameters
  optimizer.step()
```

```python
with torch.no_grad():
  y_pred = model(X_test_tensor)
  y_pred = (y_pred > 0.6).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f"Accuracy : {accuracy.item()}")
```
